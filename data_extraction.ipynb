```python
import json
import os
import pandas as pd
```

```python
path1="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path1)
Agg_state_list

#Agg_state_list (extract state)
```

```python
clm={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    p_i1=path1+i+"/"
    Agg_yr=os.listdir(p_i1)
    #Agg_yr(extract year)
    for j in Agg_yr:
      p_j1=p_i1+j+"/"
      Agg_yr_list=os.listdir(p_j1)
      #Agg_yr_list(extract json)
      for k in Agg_yr_list:
          p_k=p_j1+k
          data1=open(p_k,'r')
          d1=json.load(data1)
          #d(load the data json file and store in the variable d)
          for z in d1['data']['transactionData']:
            Name=z['name']
            count=z['paymentInstruments'][0]['count']
            amount=z['paymentInstruments'][0]['amount']
            clm['Transaction_type'].append(Name)
            clm['Transaction_count'].append(count)
            clm['Transaction_amount'].append(amount)
            clm['State'].append(i)
            clm['Year'].append(j)
            clm['Quarter'].append(int(k.strip('.json')))


agg_trans_dataframe=pd.DataFrame(clm)
```

```python
agg_trans_dataframe
```

```python
agg_trans_dataframe.to_csv('agg_trans_dataframe.csv',index=False)
```

```python
import os
import json
path2='/content/pulse/data/aggregated/user/country/india/state/'
agg_state_list2=os.listdir(path2)
agg_state_list2
#agg_state_list2(extract state name)
```

```python
clm={'State':[], 'Year':[],'Quarter':[],'Brands':[],'count':[],'Percentage':[]}
for i in agg_state_list2:
    p_i2=path2+i+"/"
    agg_yr2=os.listdir(p_i2)   #agg_yr2(extract year)
    for j in agg_yr2:
      p_j2=p_i2+j+"/"
      agg_year_list=os.listdir(p_j2) #agg_year_list(extract json file)
      for k in agg_year_list:
        p_k2=p_j2+k
        data1=open(p_k2,'r')
        d2=json.load(data1)
        try:
          for z in d2['data']['usersByDevice']:
            Brand=z['brand']
            count=z['count']
            Percentage=z['percentage']
            clm['Brands'].append(Brand)
            clm['count'].append(count)
            clm['Percentage'].append(Percentage)
            clm['State'].append(i)
            clm['Year'].append(j)
            clm['Quarter'].append(int(k.strip('.json')))
        except:
          pass
```

```python
agg_user_dataframe=pd.DataFrame(clm)
```

```python
agg_user_dataframe.to_csv('agg_user_dataframe.csv',index=False)
```

```python
import os
import json
import pandas as pd
path4="/content/pulse/data/map/user/hover/country/india/state/"
map_state_list=os.listdir(path4)
map_state_list #extract state from the path
```

```python
clm={"State": [], "Year": [], "Quarter": [], "District": [], "RegisteredUser": [], "AppOpens": []}
for i in map_state_list:
  p_i4=path4+i+"/"
  map_year_list4=os.listdir(p_i4) #extract year from the path
  for j in map_year_list4:
    p_j4=p_i4+j+'/'
    map_year_list4=os.listdir(p_j4) #extract json
    for k in map_year_list4:
      p_k4=p_j4+k
      data4=open(p_k4,'r')
      d4=json.load(data4)
      for z in d4['data']['hoverData'].items():
        district=z[0]
        registeredusers=z[1]['registeredUsers']
        appopens=z[1]['appOpens']
        clm['District'].append(district)
        clm['RegisteredUser'].append(registeredusers)
        clm['AppOpens'].append(appopens)
        clm['State'].append(i)
        clm['Year'].append(j)
        clm['Quarter'].append(int(k.strip('.json')))

map_user_dataframe=pd.DataFrame(clm)
```

```python
map_user_dataframe.to_csv('map_user_dataframe.csv',index=False)
```

```python
import os
import pandas as pd
import json
path5='/content/pulse/data/top/transaction/country/india/state/'
top_state_list=os.listdir(path5) #extract state
top_state_list
```
